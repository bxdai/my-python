```python
import os
import shutil

print(os.getcwd())
input = r'C:/Users/Bxd/Downloads/Digital Hand Atlas/JPEGimages'
output = r'C:/Users/Bxd/Downloads/Digital Hand Atlas/jpg'
print(input)
g = os.walk(input)
for path, dir_list, file_list in g:
    print(path)
    for file_name in file_list:
        filename = os.path.join(output, file_name)
        print(file_name)
        #os.rename(os.path.join(path, file_name), filename)
        shutil.copyfile(os.path.join(path, file_name),filename)

```
